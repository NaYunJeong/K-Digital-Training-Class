# 시계열 데이터 (Timeseries)

시계열 데이터 : 일정한 시간동안 수집된 일련의 순차적으로 정해진 데이터들의 집합

시계형 자료형 : timestamp, period

```python
import pandas as pd
import seaborn as sns
```

# timestamp

pd.to_datetime() : 날짜 => 시계열타입

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
flight.info()
```

```python
# flight['month']
# 영어로 되어있는 달 => 숫자로 바꾸기
months = {
    'Jan' : 1,
    'Feb' : 2,
    'Mar' : 3,
    'Apr' : 4,
    'May' : 5,
    'Jun' : 6,
    'Jul' : 7,
    'Aug' : 8,
    'Sep' : 9,
    'Oct' : 10,
    'Nov' : 11,
    'Dec' : 12
}

# map함수를 이용하면, Series를 대상으로 원하는 내용으로 적용을 할 수 있음
flight['month'] = flight['month'].map(months)
flight
```

```python
# 열을 하나 추가
flight.insert(2, 'day', 1)    # 원하는 위치에 열을 추가하고 싶을 때 : insert(위치, 컬럼명, 값)
flight.head()
```

```python
# year, month, day => 날짜
flight['date'] = pd.to_datetime(flight[['year', 'month', 'day']])
flight
```

# Timestamp

Python의 datetime과 유사

특정날짜를 표현 가능 O

```python
# 위치로 (year, month, day, hour, minute, second)
christmas = pd.Timestamp(2024, 12, 25, 14, 13, 21)
christmas
```

```python
# 파라미터명 명시해서
christmas2 = pd.Timestamp(day = 25, hour = 12, month = 12, minute = 12,
                          second = 12, year = 2024)
christmas2
```

```python
solo = pd.Timestamp(2022, 12, 26, 12, 30, 19)
print('솔로된 날짜 : ', solo)
print('솔로된 연도 : ', solo.year)
print('솔로된 월 : ', solo.month)
print('솔로된 일자 : ', solo.day)
print('솔로된 시간 : ', solo.hour)
print('솔로된 분 : ', solo.minute)
print('솔로된 초 : ', solo.second)
print('솔로된 요일 : ', solo.dayofweek) # 요일] 월 : 0 ~ 일 : 6
print('일년 내 며칠 째에 하는지 : ', solo.dayofyear)
print('해당하는 달이 며칠까지 있는지 : ', solo.daysinmonth)
print('몇 분기에 해당 하는지', solo.quarter)
print('몇 주째에 해당 하는지', solo.week)
```

```python
print(solo.date())  # 날짜
print(solo.time())  # 시간
print(pd.Timestamp.combine(solo.date(), solo.time()))
# combine() 함수 : date() 함수와 time() 함수를 합쳐서 날짜/ 시간 객체 생성
```

# Period

pd.Timestamp()와 pd.Period()의 차이
Timestamp는 하나의 시점을 뜻하고, Period는 지정한 날짜의 시작지점부터 종료지점까지의 범위를 포괄함

```python
p = pd.Period('2024-08-15')
p
print(p.start_time)
print(p.end_time)
```

```python
dates = ['2024-08-15', '2024-10-15', '2024-12-15']
datesD = pd.to_datetime(dates)
datesD
```

```python
# timestamp => period
forDay = datesD.to_period(freq='D')
forDay
```

```python
forMonth = datesD.to_period(freq='M')
forMonth
```

```python
forYear = datesD.to_period(freq='A')
forYear
```

# Timestamp 배열

date_range() 함수

```python
pd.date_range('2024-08-02', periods=12, freq='2D')    # periods(구간) 12개  # freq 2D (2일씩 늘려가면서)
```

```python
a = pd.date_range(start = '2024-08-02',     # 시작 날짜
                  end = None,               # 끝 날짜 (None일 경우 생략 가능)
                  periods = 8,              # 생성할 timestamp 갯수
                  freq = '2M'               # 시간 간격
                  )
a
```

- M : 각 달의 마지막날짜
- MS : 각 달의 첫 날
- D : 일자별

# Period 배열

period_range() 함수

```python
# 두 달 단위
twoM = pd.period_range('2009-07-14', periods=10, freq='2M')
twoM
```
