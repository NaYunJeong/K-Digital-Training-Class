# 3. 기초 통계 정보 확인하기

# 3-1. 기초통계량 : describe()

총 데이터 수(count), 평균(mean), 표준편자(std), 4분위수(25,50,75%),
최대/최소(max/min), 중앙값(median)

중복이 없는 고유 데이터값(unique), 가장 값이 많은 데이터(Top),
최빈 데이터의 실제 수(Freq)
(ex: 1000개의 데이터 성별 중 남 700 , 여 300명 일 때,
Top값은 남성 / Freq값은 700)

# 평균값 vs 중앙값

평균값 (mean) : 값들을 다 더한 후에 값의 갯수로 나눈 값
=> ex) 5명의 월급이 각각 100, 200, 150, 170, 300
=> 평균값 : 184만원

중앙값 (median) : 값들을 크기 순으로 나열한 후에 가장 중앙에 위치하는 값
=> ex) 5명의 월급이 각각 100, 200, 150, 170, 300
=> 중앙값 : 100 150 170 200 300 => 가장 중앙 => 170만원

# 3-5. 상관계수(문자열은 제외) : corr()

상관관계는 두 변수간의 상관성에 대한 정보로 -1 ~ 1 사이의 값을 갖게 됨

- 변수 간 상관관계 : 변수간 흐름이 얼마나 비슷한가를 나타내주는 척도
- a가 증가하면 b도 증가하는가 / 감소하는가
- or 증가의 성향이 얼마나 비슷한가

를 알 수 있음

1에 가까울수록 두 지표는 비슷하게 증가 or 감소

-1에 가까울수록 하나 증가 / 하나 감소

0에 가까울수록 두 지표는 서로 관계가 없다

꼭 인과관계를 의미하지는 않음

이 지표는 독립변수와 종속변수(독립변수에 영향을 받는 변수) 로 만들어져야 함

데이터들의 특징을 이해하는데 있어서
(EDA과정 - Exploratory Data Analysis, 탐색적 자료 분석) 사용해 볼 만 하다

corr() 함수는 문자열을 제외한 변수간 matrix를 생성하고, 각 쌍들의 상관계수를 반환

DataFrame, Matrix 둘다 2차원 형태의 데이터 테이블 형태

DataFrame은 각 요소가 다른 자료형이어도 되지만,

Matrix는 각 요소의 자료형을 일치시켜야 한다는 차이점



```python
i.corr(numeric_only=True)
```
