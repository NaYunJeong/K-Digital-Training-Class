# 1. 열 순서 변경 : DataFrame객체[열이름의 리스트]

```python
titanic = sns.load_dataset('titanic')
# titanic
df = titanic.loc[0:4, 'age':'fare']
df
```

```python
# 보기 쉽게 컬럼명을 정렬해서 열 순서 바꾸자!!
columns = list(df.columns)
print(columns)

srt_col = sorted(columns) # abc순서 정렬
print(srt_col)

srt_col2 = sorted(columns, reverse=True) # cba순서 정렬
print(srt_col2)
```

# 2. 열 분리 : Series.str.get(인덱스)

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
# 월에 영어값인거 => 각각 숫자로 변경해서 데이터에 적용
months = {
    'Jan' : 1,
    'Feb' : 2,
    'Mar' : 3,
    'Apr' : 4,
    'May' : 5,
    'Jun' : 6,
    'Jul' : 7,
    'Aug' : 8,
    'Sep' : 9,
    'Oct' : 10,
    'Nov' : 11,
    'Dec' : 12
}

flight['month'] = flight['month'].map(months)
flight
```

```python
# month 컬럼 옆에 day 컬럼을 만들고 컬럼값을 모두 15
flight.insert(2, 'day', 15)
flight.head()
```

```python
# year, month, day 열을 합친 날짜데이터를 'date' 컬럼에 추가
flight['date'] = pd.to_datetime(flight[['year', 'month', 'day']])
flight.head()
```

```python
flight['date'] = flight['date'].astype('str')
dates = flight['date'].str.split('-')
```

```python
# dates변수는 리스트[연, 월, 일]를 개별요소로 갖는 시리즈 객체 형태
# str.get()함수를 이용하면 각 행별로 인덱싱이 가능
# str.get() : 리스트로 변환 후 해당 인덱스 위치의 요소를 반환
flight['y'] = dates.str.get(0)
flight['m'] = dates.str.get(1)
flight['d'] = dates.str.get(2)
flight.head()
```

# 2-1. 매핑함수를 이용해서 열 분리

매핑함수(람다함수)와 for문을 활용해서 위와 같은 결과 도출해보기

```python
flight['date'] = flight['date'].astype('str')
col = ['y', 'm', 'd']
for i in range(len(col)):
  flight[col[i]] = flight['date'].apply(lambda x: x.split('-')[i])

flight.head()

# series객체. str.get()을 기억하자!!
```
